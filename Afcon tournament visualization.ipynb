```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as snb
import networkx as nx
```

```python
data = pd.read_csv('international-africa-cup-of-nations-matches-2019-to-2019-stats.csv')
```

```python
data.head().T
```

Here we can view the  home team and the sum of goals scored by the home team.

```python
a=data.groupby(['home_team_name']).sum().reset_index()
fig, ax = plt.subplots(figsize=(30, 6))
snb.barplot(x="home_team_name",y="home_team_goal_count",data = a,ax=ax,)
plt.title("home_team_goal_count")
```

Here we can view the  away team and the sum of goals scored by the away team.

```python
a=data.groupby(['away_team_name']).sum().reset_index()
fig, ax = plt.subplots(figsize=(30, 6))
snb.barplot(x="away_team_name",y="away_team_goal_count",data = a,ax=ax,)
plt.title("away_team_goal_count")
```

```python
def lab(data):
    if data["home_team_goal_count"] > data["away_team_goal_count"]:
        return "HOME TEAM WIN"
    elif data["away_team_goal_count"] > data["home_team_goal_count"]:
        return "AWAY TEAM WIN"
    elif data["home_team_goal_count"] == data["away_team_goal_count"]:
        return "DRAW"
```

```python
def home_win(data):
    if data["home_team_goal_count"] > data["away_team_goal_count"]:
        return 1
    elif data["away_team_goal_count"] > data["home_team_goal_count"]:
        return 0
    elif data["home_team_goal_count"] == data["away_team_goal_count"]:
        return 0
```

```python
def away_win(data):
    if data["home_team_goal_count"] > data["away_team_goal_count"]:
        return 0
    elif data["away_team_goal_count"] > data["home_team_goal_count"]:
        return 1
    elif data["home_team_goal_count"] == data["away_team_goal_count"]:
        return 0
```

```python
data_["outcome_side"] = data_.apply(lambda data:lab(data),axis = 1)
```

```python
data_["home_win"] = data_.apply(lambda data:home_win(data),axis = 1)
data_["away_win"] = data_.apply(lambda data:away_win(data),axis = 1)
```

```python
def match(data):
    return str(str(data['home_team_name']) +' vs ' + str(data['away_team_name']))
```

```python
data_["match"] = data_.apply(lambda data:match(data),axis = 1)
```

```python
data_.plot(x='match',y=['home_team_possession','away_team_possession'], kind="barh",figsize=(10,30))
```
